```python
import numpy as np
n=1000
vals = np.random.logistic(loc=5, scale=0.5, size=n)
print(vals)
```

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

```python
from scipy.stats import logistic
import numpy as np
import matplotlib.pyplot as plt

#logistic mu=5 sigma=0.5
x= np.arange(-1,10,0.001)
plt.plot(x, logistic.pdf(x, loc=5, scale=0.5))
plt.title('Funcin de densidad logistic')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```
